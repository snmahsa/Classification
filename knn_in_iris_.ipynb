```python
import numpy as np
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
#Import Dataset
dataset = pd.read_csv('/content/drive/MyDrive/IRIS.csv')
dataset
```

```python
x_data = dataset[['sepal_length','sepal_width','petal_length','petal_width']]
y_data = dataset[['species']]
```

```python
x_data
```

```python
y_data
```

```python

# Assuming 'target' is the name of the target variable column in your dataset
target = y_data

# Create an instance of LabelEncoder
label_encoder = LabelEncoder()

# Fit the label encoder to the target variable
label_encoder.fit(target)

# Transform the target variable using label encoding
encoded_target = label_encoder.transform(target)

print(encoded_target)
```

```python
# Add the new column to the dataset
new_dataset = x_data.copy()
new_dataset['num_species'] = encoded_target
new_dataset
```

```python
#Split
x_train , x_test , y_train , y_test = train_test_split(x_data , encoded_target, test_size=0.2,random_state=42)
```

```python
x_train.shape, x_test.shape ,y_train.shape, y_test.shape
```

```python
#Model
model = KNeighborsClassifier(n_neighbors=7)
model.fit(x_train,y_train)
```

```python
#Prediction
y_pred = model.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
accuracy
```
